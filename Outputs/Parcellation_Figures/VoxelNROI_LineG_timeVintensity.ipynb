```python
###1) VOXEL PRE PARCELLATION

#Importing packages
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step 1: Load the fMRI NIfTI file
nifti_file_path = '/Users/judy/Desktop/Research/Code/test_sub.nii' 
nifti_img = nib.load(nifti_file_path)

# Step 2: Extract the voxel data
voxel_data = nifti_img.get_fdata()  # Get the voxel data as a NumPy array

# Step 3: Specify the coordinates of the voxel (x, y, z)
voxel_coords = (30, 40, 20)  # Adjust these coordinates as needed

# Step 4: Extract the time series for the specified voxel
time_series = voxel_data[voxel_coords]  # This will give you the time series

# Step 5: Remove NaN values if present
time_series = time_series[~np.isnan(time_series)]
```

```python
## Graphing
# Step 6: Plot the line graph
time_x = np.arange(1,len(time_series)+1)
intensity_y = time_series

plt.plot(time_x, intensity_y, color='r')

plt.xlabel('Time (327 time points total)')
plt.ylabel('Intensity of Activation')
plt.title(f'Intensity of Activation of one Voxel at {voxel_coords} Pre-parcellation, over Time')

plt.show()
```

```python
###2) ROI POST PARCELLATION

# Step 1: Load the Parcellated File (already nmuPy.ndarray with 327 rows (time), 100 columns(ROIs))
file_path_roi = '/Users/judy/parcellated_test_sub.txt'
roi_data = np.loadtxt(file_path_roi)

# Step 2: Specify the coordinate of the ROI (x of graph)
roi_coord = (99)  # Adjust these coordinates as needed

# Step 4: Extract the time series for the specified voxel
time_series_roi = roi_data[:, roi_coord]  # This will give you the time series

# Step 5: Remove NaN values if present
time_series_roi = time_series_roi[~np.isnan(time_series_roi)]
```

```python
## Graphing
# Step 6: Plot the line graph
time_x_roi = np.arange(1,len(time_series_roi)+1)
intensity_y_roi = time_series_roi

plt.plot(time_x_roi, intensity_y_roi, color='g')

plt.xlabel('Time (327 time points total)')
plt.ylabel('Intensity of Activation')
plt.title(f'Intensity of Activation of ROI #{roi_coord} Post-parcellation, over Time - TEMPORAL FILTER')

plt.show()
```

```python
## Graphing
# Step 6: Plot the line graph
time_x_roi = np.arange(1,len(time_series_roi)+1)
intensity_y_roi = time_series_roi

plt.hist(intensity_y_roi, 100)

plt.xlabel('Time (327 time points total)')
plt.ylabel('Intensity of Activation')
plt.title(f'Intensity of Activation of ROI #{roi_coord} Post-parcellation, over Time')

plt.show()
```
